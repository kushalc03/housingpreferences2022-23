```python
import csv
import matplotlib
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('housing_prefs.csv')
df
```

```python
df.mean(axis=0)
```

```python
athlete = df[df['Are you an athlete?'] == 'Yes'].drop(columns=['Blocking group size']).mean(axis=0)
non_athlete = df[df['Are you an athlete?'] == 'No'].drop(columns=['Blocking group size']).mean(axis=0)

df = pd.DataFrame({'Athlete': athlete,
                   'Non-Athlete': non_athlete})
ax = df.plot(kind='bar', title='Rankings for Athletes and Non-Athletes', color=["#D2232A", "000000"], fontsize=10, width=0.9, figsize=(15,8))
for p in ax.patches:
    ax.annotate(np.round(p.get_height(),decimals=2), (p.get_x()+p.get_width()/2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=10)
ax.legend(loc=1, prop={'size': 15})
ax.set_xlabel("House", fontsize=15)
ax.set_ylabel("Ranking", fontsize=15)
```

```python
ax.get_figure().savefig('athlete.png')
```
